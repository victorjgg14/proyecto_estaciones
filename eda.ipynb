```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'tabla_estaciones.csv'
df = pd.read_csv(url)
```

```python
df.head(10)
```

```python
df.groupby('name')[['lifts_total', 'runs_total', 'km_total']].mean(numeric_only=True).sort_values(by='km_total', ascending=False)
```

```python
top10_estaciones_km = df[['name','km_total']].sort_values(by='km_total', ascending=False).head(10)
```

# Análisis de frecuencias: 
- frecuencia de valores únicos en las variables categóricas ('trineos', 'esqui_de_fondo', 'snowpark').

```python
# Calcular la frecuencia de valores únicos en las variables categóricas

categorical_columns = ['trineos', 'esqui_de_fondo', 'snowpark']
frequency_counts = {}

for column in categorical_columns:
    
    frequency_counts[column] = df[column].value_counts()

# Mostrar las frecuencias

for column, counts in frequency_counts.items():
    
    print(f"Frecuencia de valores en la columna '{column}':")
    print(counts)
    print()
```

# Análisis comparativo: 
- comparar diferentes variables entre sí para encontrar posibles relaciones o patrones. Por ejemplo, comparar el número total de carreras ('runs_total') con el número total de ascensores ('lifts_total') para ver si hay alguna relación entre ellos.

```python

# Calcular la correlación entre 'runs_total' y 'lifts_total'
correlation = df['runs_total'].corr(df['lifts_total'])
print(f"La correlación entre el número total de pistas y el número total de telesillas es: {correlation}")
```


- Tenemos una correlación de 0.827 entre el número total de pistas y el número total de telesillas, lo que indica una correlación positiva fuerte entre estas dos variables.

- Esto significa que a medida que aumenta el número total de pistas, tiende a aumentar también el número total de telesillas, y viceversa. Esta información puede ser útil para tomar decisiones sobre la planificación y gestión de recursos en una estación de esquí, por ejemplo, al considerar la capacidad de los telesillas en relación con la cantidad de pistas disponibles.

```python

# Graficar un diagrama de dispersión para comparar 'runs_total' con 'lifts_total'
plt.scatter(df['lifts_total'], df['runs_total'])
plt.title('Comparación entre número total de pistas y número total de telesillas')
plt.xlabel('Número total de telesillas')
plt.ylabel('Número total de pistas')
plt.grid(True)
plt.show()
```

```python
# Calcular la correlación entre 'runs_total' y 'lifts_total'

correlation = df['km_total'].corr(df['runs_total'])

print(f"La correlación entre el número total de km totales y el número total de pistas es: {correlation}")
```

```python
# Graficar un diagrama de dispersión para comparar 'km_total' con 'runs_total'

plt.scatter(df['km_total'], df['runs_total'])

plt.title('Comparación entre número total de km y número total de pistas')

plt.xlabel('Número total de km')

plt.ylabel('Número total de pistas')

plt.grid(True)

plt.show()
```
